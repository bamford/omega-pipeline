```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
import emcee
import triangle
```

```python
wlHa=6562.8
wlNIIa =6548.1
wlNIIb=6583.5

par = ['continuum', 'redshift', 'flux Ha', 'flux NII']
ndim_fixha = 4

nwalkers = 300
nburn = 100
nsamp = 200
```

```python
def gaussian(x, ampl, centre, sigma):
    # with unit area
    return (ampl/(sigma*sqrt(2*pi))) * np.exp(-(x-centre)**2/(2.0*sigma**2))
```

```python
def fixha_model(x, p):
    model = p[0] + gaussian(x, p[2], (1+p[1])*wlHa, 7)
    model += gaussian(x, p[3]/3.06, (1+p[1])*wlNIIa, 7) + gaussian(x, p[3], (1+p[1])*wlNIIb, 7)
    return model
```

```python
def lnprob_fixha(p, y, x, icov, max_fha, zmin, zmax):
    # Here we set the hard priors:
    if (p[3] < 0) or (p[1] > zmax) or (p[1] < zmin) or ((p[3]/p[2]) > 3):
        return -np.inf
    model = fixha_model(x, p)
    chisq = ((model - y)**2 * icov).sum()
    return -chisq/2.0
```

```python
def lnprob_fixha_bruno(x, mu,xmeans,icov,max_fha,zmin,zmax):
    model = np.zeros(len(mu))
    diff = np.zeros(len(mu))

    # Fix for different treatment of icov
    icov = np.diag(icov)
    
    ###Here we define our model:
    for i in range(len(xmeans)-1):
	model[i] = x[0] + gaussian(xmeans[i],x[2],(1+x[1])*wlHa,7)  +\
	(1)*gaussian(xmeans[i],x[3]/3.06,(1+x[1])*wlNIIa,7)  + gaussian(xmeans[i],x[3],(1+x[1])*wlNIIb,7)
   ###Here we set the priors:
    for j in range(len(model)):
        diff[j] = model[j] - mu[j]
	#if (x[3] < 0):
        if (x[3] < 0) or (x[1] > zmax) or (x[1] < zmin) or ((x[3]/x[2]) > 3):
	    #diff[j]=10**300
            return -np.inf
    return -np.dot(diff,np.dot(icov,diff))/2.0
```

```python
def get_data_bruno(glx):
    import math
    filename='gal_'+str(glx)+'_aper_5.txt'
    f = np.loadtxt(filename,dtype='f')
    xmeans = f[:,0]
    ymeans = f[:,1]
    error_flux_phot=f[:,2]
    error_flux_zeropointfit=f[:,3]
    error_flux=f[:,4]
  
 ## Getting rid of the nan values... 
    for ii in range(0,len(ymeans)):
        if (math.isnan(ymeans[ii]) == True) or (math.isnan(error_flux[ii]) == True) or \
(math.isnan(-1*ymeans[ii]) == True) or (math.isnan(-1*error_flux[ii]) == True):
            xmeans[ii]=-501
	    ymeans[ii]=-501
            error_flux[ii]=-501
	if error_flux[ii] == 0:
	    error_flux[ii] =1

    a = np.where((xmeans > -500) & (ymeans > -500))
    xmeans = xmeans[a[0]]
    ymeans = ymeans[a[0]]
    error_flux = error_flux[a[0]]
    xmin=min(xmeans)
    xmax=max(xmeans)
    ymin=min(ymeans)
    ymax=max(ymeans)

    error2 = (error_flux)*(error_flux)
    cov=np.diag((error2))
    
    icov = np.linalg.inv(cov)

    zmax = (xmax+20)/wlHa -1
    zmin = (xmin-20)/wlHa -1

    return xmeans, ymeans, error_flux, icov, xmin, xmax, ymin, ymax, zmin, zmax
```

```python
def get_data(glx):
    filename='gal_'+str(glx)+'_aper_5.txt'
    f = np.loadtxt(filename, dtype='f')
    xmeans, ymeans, error_flux_phot, error_flux_zeropointfit, error_flux = f[:,:5].T

    ## Getting rid of the nan values...
    ok = np.logical_not(np.isnan(ymeans) | np.isnan(error_flux))
    
    error_flux[error_flux == 0] = 1.0  # ARBITRARY?!?
        
    xmeans = xmeans[ok]
    ymeans = ymeans[ok]
    error_flux = error_flux[ok]
    xmin = np.min(xmeans)
    xmax = np.max(xmeans)
    ymin = np.min(ymeans)
    ymax = np.max(ymeans)
    icov = 1/error_flux**2
    zmax = (xmax+20)/wlHa - 1
    zmin = (xmin-20)/wlHa - 1
    return xmeans, ymeans, error_flux, icov, xmin, xmax, ymin, ymax, zmin, zmax
```

```python
def flatten_without_burn(sampler, nburn):
    c = sampler.chain
    if c.ndim == 4:
        c = c[0]
    c = c[:, nburn:]
    return c.reshape((np.product(c.shape[:-1]), c.shape[-1]))

def weight_without_burn(sampler, nburn):
    c = sampler.lnprobability
    if c.ndim == 3:
        c = c[0]
    c = c[:, nburn:]
    w = np.exp(c.reshape(np.product(c.shape)))
    return w / w.sum()
```

```python
def init_p0_fixha_bruno():
    ### Defining the initial shifts for the walkers in each parameter

    c0_shift=100
    z_shift=0.01
    aha_shift=100
    sha_shift=100
    anii_shift=100
    
    shifts_fixha=[c0_shift,z_shift,aha_shift,anii_shift]

    nwalkers = 300
    
### Initial value of each parameter    
    init_fha=(ymeans.max()-ymeans.min())*sqrt(2*pi)*7### 7 is the guessed sigma of Ha
    max_fha = ymeans.max()*sqrt(2*pi)*7  
    init_z=xmeans[np.where(ymeans == ymeans.max())]/wlHa - 1


    if wlNIIb*(1+init_z[0]) > xmeans[len(xmeans)-1]:
        
        init_fnii=0
    else:
        init_fnii=((ymeans[np.where(xmeans > wlNIIb*(1+init_z))[0][0]] - ymeans.min())*sqrt(2*pi)*7)/3.06


    ## RUN MCMC FOR FIXED HA FIXED C1

    p0_0_fixha = [ymeans[0],zmin,init_fha-500,init_fha-500]

    p0_fixha = np.random.rand(ndim_fixha * nwalkers).reshape((nwalkers, ndim_fixha))
    
    for i in range(ndim_fixha):
        p0_fixha[:,i] = p0_fixha[:,i] - p0_fixha[:,i]  + p0_0_fixha[i] + shifts_fixha[i]*np.random.rand(nwalkers)

    return p0_fixha
```

```python
def init_p0_fixha():
    # very broad uniform initial distributions
    c0_init = np.random.uniform(ymin, ymax, nwalkers)
    z_init = np.random.uniform(zmin, zmax, nwalkers)
    fha_init = (np.random.uniform(ymin, ymax, nwalkers) - ymin) * sqrt(2*pi) * 7
    fnii_init = (np.random.uniform(ymin, ymax, nwalkers) - ymin) * sqrt(2*pi) * 7
    return np.transpose([c0_init, z_init, fha_init, fnii_init])
```

```python
def compare_init(i):
    # compare Bruno and Steven's initial parameters
    a = init_p0_fixha_bruno()[:,i]
    b = init_p0_fixha()[:,i]
    plt.hist(b, label='steven')
    plt.hist(a, label='bruno', alpha=0.5)
    plt.legend()
    plt.show()
```

```python
xmeans, ymeans, error_flux, icov, xmin, xmax, ymin, ymax, zmin, zmax = get_data(glx)

p0_fixha = init_p0_fixha()

# value doesn't actually matter here
max_fha = ymax*sqrt(2*pi)*7
```

```python
sampler_fixha = emcee.EnsembleSampler(nwalkers, ndim_fixha, lnprob_fixha,
                                      args=[ymeans, xmeans, icov, max_fha, zmin, zmax], threads=4)
r = sampler_fixha.run_mcmc(p0_fixha, nburn+nsamp)
```
